```python
import pandas as pd
```

```python
income_df = pd.read_csv("final_income.csv")
income_df
```

```python
income_df2 =income_df.drop("Avg Income (12-17)",axis=1)
income_df2=income_df2.set_index("State").stack().reset_index(name='Income')\
    .rename(columns={'level_1':'Year'})
income_df2
```

```python
bachelors_df = pd.read_csv("bachelors-degree.csv")
bachelors_df
```

```python
bachelors_df=bachelors_df.set_index("State").stack().reset_index(name='Bachelors')\
    .rename(columns={'level_1':'Year'})
bachelors_df
```

```python
income_df2['Year']=income_df2['Year'].astype(int)
```

```python
bachelors_df['Year']=bachelors_df['Year'].astype(int)
```

```python
merge_df1 = bachelors_df.merge(income_df2, how='right')
```

```python
merge_df1
```

```python
merge_df1.to_csv(r'../CleanData/merge_bi.csv',index=False)
```
